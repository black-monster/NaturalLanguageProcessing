```python
from nltk.corpus import gutenberg
```

```python
gutenberg.fileids()
```

```python
for file_id in gutenberg.fileids():
    num_chars = len(gutenberg.raw(file_id))
    num_words = len(gutenberg.words(file_id))
    num_sents = len(gutenberg.sents(file_id))
    num_vocab = len(set([w.lower() for w in gutenberg.words(file_id)]))
    print (int(num_chars/num_words), int(num_words/num_sents), int(num_words/num_vocab), file_id)
    
""". Observe that average word length appears to be a
general property of English, since it has a recurrent value of 4. (In fact, the average word
length is really 3, not 4, since the num_chars variable counts space characters.) By contrast average sentence length and lexical diversity appear to be characteristics of particular authors.
"""
```

## Brown Corpus

```python
from nltk.corpus import brown
```

```python
# nltk.download('brown')
```

```python
brown.categories()
```

```python
brown.words(categories='news')
```

```python
news_text = brown.words(categories='news')
fdist = nltk.FreqDist([w.lower() for w in news_text])
```

```python
fdist
```

```python
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
    print (m + ":", fdist[m])
```

```python
cfd = nltk.ConditionalFreqDist( (genre, word) 
                               for genre in brown.categories()
                              for word in brown.words(categories=genre))
```

```python
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
```

```python
 modals = ['can', 'could', 'may', 'might', 'must', 'will']
```

```python
cfd.tabulate(conditions=genres, samples=modals)
```

## Inaugral address corpus

```python
from nltk.corpus import inaugural
```

```python
inaugural.fileids()
```

```python
[fileid[:4] for fileid in inaugural.fileids()]
```

```python
inaugural.words
```

```python
cfd = nltk.ConditionalFreqDist( (target, fileid[:4])
                               for fileid in inaugural.fileids()
                               for w in inaugural.words(fileid)
                               for target in ['america', 'citizen']
                               if w.lower().startswith(target))

cfd.plot()
```

```python
from nltk.corpus import udhr
languages = ['Chickasaw', 'English', 'German_Deutsch','Greenlandic_Inuktikut', 'Hungarian_Magyar', 'Ibibio_Efik']
```

```python
cfd = nltk.ConditionalFreqDist((lang, len(word)) for lang in languages for word in udhr.words(lang + '-Latin1'))
cfd.plot(cumulative=True)
```
